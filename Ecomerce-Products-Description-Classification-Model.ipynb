<h3>Ecommerce Products Classification From thier Descriptions <a href='https://www.kaggle.com/datasets/darshakofficial/ecommerce-data'>Ecommerce_data.csv Dataset Link (Kaggle)</a></h3>

```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/Ecommerce_data.csv")
df
```

```python
df.label.value_counts()
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.text)
    return " ".join(filtered_tokens)    
```

```python
df["Preprocessed_Text"] = df.Text.apply(preprocess)
```

```python
df["label_num"] = df.label.map({
    "Household":0,
    "Electronics":1,
    "Clothing & Accessories":2,
    "Books":4
})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    df.Preprocessed_Text,
    df.label_num,
    test_size=0.2,
    random_state=2022,
    stratify=df.label_num
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
clf = Pipeline([
    ('vectorizer_tfidf',TfidfVectorizer()),
    ('Random_Forest_Classifier',RandomForestClassifier())
])
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```
