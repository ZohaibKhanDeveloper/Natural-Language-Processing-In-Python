<h2>Spam Email Classification Model Implementation</h2>

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Datasets/spam.csv')
df
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr,xts,ytr,yts = train_test_split(df.Message,df.spam,test_size=0.2)
```

```python
xtr.shape
```

```python
xts.shape
```

```python
type(xtr)
```

```python
xtr[:4]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
xtr_cv = v.fit_transform(xtr.values)
```

```python
xtr_cv.shape
```

```python
v.get_feature_names_out()
```

```python
v.vocabulary_
```

```python
xtr_np = xtr_cv.toarray()
```

```python
xtr_np[0]
```

```python
np.where(xtr_np[0]!=0)
```

```python
xtr[:4]
```

```python
xtr[:4][1364]
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(xtr_cv,ytr)
```

```python
xts_cv = v.transform(xts)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = model.predict(xts_cv)
```

```python
print(classification_report(yts,y_pred))
```

<h2>Model Implementation (Easy Method using Pipeline)</h2>

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('vctorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(xtr,ytr)
```

```python
y_pred = clf.predict(xts)
```

```python
print(classification_report(yts,y_pred))
```

```python
clf.predict(emails)
```
