<h3>Fake News Classification Model Using Word Embeddings <a href='https://www.kaggle.com/datasets/shubh0799/fake-news'>Fake or Real News Dataset Link (Kaggle)</a></h3>

```python
import pandas as pd
import spacy
```

```python
df = pd.read_csv("Datasets/fake_or_real_news.csv")
```

```python
df
```

```python
df.label.value_counts()
```

```python
len(df['text'][0])
```

```python
df["label_num"] = df.label.map({
    "FAKE":0,
    "REAL":1
})
```

```python
df
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
df["vector"] = df["text"].apply(lambda x: nlp(x).vector)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    df.vector.values,
    df.label_num,
    test_size=0.2,
    random_state=2022
)
```

```python
import numpy as np
```

```python
x_train_2d = np.stack(x_train)
x_test_2d = np.stack(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_train_embed = scaler.fit_transform(x_train_2d)
scaled_test_embed = scaler.transform(x_test_2d)
```

```python
clf = MultinomialNB()
clf.fit(scaled_train_embed,y_train)
```

```python
y_pred = clf.predict(scaled_test_embed)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier(n_neighbors=5,metric='euclidean')
```

```python
clf.fit(x_train_2d,y_train)
```

```python
y_pred = clf.predict(x_test_2d)
```

```python
print(classification_report(y_test,y_pred))
```
