<h3>Ecommerce Products Classification Using FastText (Preprocessing) <a href='https://www.kaggle.com/datasets/darshakofficial/ecommerce-data'>Ecommerce_data.csv Dataset Link (Kaggle)</a></h3>

```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/Ecommerce_Data.csv")
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.label = df.label.replace("Clothing & Accessories","Clothing_Accessories")
```

```python
df
```

```python
df.label.value_counts()
```

```python
df.label = "__label__"+df.label.astype(str)
```

```python
df
```

```python
df["category_description"] = df["label"]+" "+df["Text"]
```

```python
df
```

```python
import re
def preprocess(text):
    text = re.sub(r"[^\w\s]"," ",text)
    text = re.sub(r" +"," ",text)
    return text.strip().lower()
```

```python
df['category_description'] = df['category_description'].map(preprocess)
df = df['category_description']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(df,test_size=0.2)
```

```python
train.shape, test.shape
```

```python
train.to_csv("ecommerce.train",columns=["category_description"],index=False,header=False)
```

```python
test.to_csv("ecommerce.test",columns=["category_description"],index=False,header=False)
```
